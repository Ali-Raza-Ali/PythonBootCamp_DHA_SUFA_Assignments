# Importing os library

```python
import os
```

```python
dir_dict = {0:{'Folder':None,'SubFolder':None,'File':None,'Size':None}}
            
count = 0

for folders, subfolders, filenames in os.walk('/Users/aliraza/Documents/Semester 4'):
    dir_dict[count]['Folder'] = folders
    
    
    for subfolder in subfolders:
        dir_dict[count]['SubFolder'] = subfolder
    
    for file in filenames:
        dir_dict[count]['File'] = file
        dir_dict[count]['Size'] = os.stat(folders+'/'+file)
        dir_dict[count]['Size'] = dir_dict[count]['Size'].st_size
    count = count + 1
    dir_dict[count] = {}
        
#     print("\n")
```

#### Dictionary  Output

```python
dir_dict
```

## Importing pandas to convert dictionary to dataframe 

```python
import pandas as pd
```

### Converting dictionary to dataframe and sorting it

```python
my_files = pd.DataFrame(dir_dict).transpose()

my_files = my_files.sort_values('Size',ascending=False)
my_files.drop(58,inplace = True)
```

### Sorted Output:

```python
print("Here NaN in column value means the we are still in the same previously occured folder")
display(my_files)
```
