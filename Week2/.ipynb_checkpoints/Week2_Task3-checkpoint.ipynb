```python
from bs4 import BeautifulSoup
import requests
import json
```

```python
res = requests.get('https://directory.ntschools.net/api/System/GetAllSchools')
```

```python
res.status_code
```

```python
res.content
```

```python
soup = BeautifulSoup(res.content,'html.parser')
```

```python
fetched_data = json.loads(res.content)
fetched_data = fetched_data[0:50] #For first 50 schools
fetched_data
```

```python
for i in range(len(fetched_data)):
    print(fetched_data[i]['itSchoolCode'])
```

```python
url2 = 'https://directory.ntschools.net/api/System/GetSchool?itSchoolCode='
Schools_data = []
for i in range(len(fetched_data)):
    Schools_data.append(requests.get((url2 + fetched_data[i]['itSchoolCode'])))
```

```python
Schools_data
```

```python
Schools_content = {}
for i in range(len(Schools_data)):
    Schools_content[i] = Schools_data[i].content
```

```python
JSONS = {}
for i in range(len(Schools_content)):
    JSONS[i] = json.loads(Schools_content[i])
JSONS
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(JSONS).transpose()
df=df[['name','physicalAddress','schoolManagement','mail','telephoneNumber']]
df.columns = ['SchoolName','Address','SchoolManagement','E-Mail','Telephone#']
df.to_csv('50Schools.csv')
```
